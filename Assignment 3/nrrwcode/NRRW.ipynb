```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress


## Do I have to also log the RMS ERROR TOO
filename = 'nrrw2.txt'
data = np.loadtxt(filename,delimiter=' ')

N= data[:,0]
Rn_squared = data[:,1]
rms_Rn2 = data[:,5]
rms_Rn = data[:,6]
log_N = np.log(N)
log_Rn2 = np.log(Rn_squared)

slope, intercept, r_value, p_value, std_err = linregress(log_N, log_Rn2)


print(f'Slope of the linearized data: {slope:.5f}')
plt.errorbar(log_N, log_Rn2, yerr= rms_Rn2/Rn_squared,fmt='o', label=r'log( $\langle R_N^2 \rangle$) vs log(N)', capsize=5)
plt.plot(log_N, slope*log_N + intercept, color = 'red',label=f'Fit: slope = {slope:.5f}')
plt.title(r'Log-Log Plot of $\langle R_N^2 \rangle$ vs N for Self Avoiding Walk where sample size S=90000')
plt.xlabel(r'log(N)')
plt.ylabel(r'log( $\langle R_N^2 \rangle$)')
plt.legend()
plt.tight_layout()
plt.show()

```

```python

N = data[:, 0]        
SAWs = data[:, 2]     
Xvalue = 1 / N
Yvalue = SAWs ** Xvalue
slope, intercept = np.polyfit(Xvalue, Yvalue, 1)
print(intercept)
Yfit = slope * Xvalue + intercept
plt.scatter(Xvalue, Yvalue, marker='o', label=r'${1/N}$ vs $C_N^{\frac{1}{N}}$)', color='red')
plt.plot(Xvalue, Yfit, label=f'Fitted Line: y = {slope:.5f}x + {intercept:.5f}', color='blue')
plt.xlabel(r'$\frac{1}{N}$')
plt.ylabel(r'$C_N^{\frac{1}{N}}$')
plt.title(r'Fitted Line for $\frac{1}{N}$ vs $C_N^{\frac{1}{N}}$)')
plt.legend()
plt.show()
```
