```python
import numpy as np 
import matplotlib.pyplot as plt 

filename = 'walks.txt'
data = np.loadtxt(filename)

N = data[:, 0]
Rn_squared= data[:, 1]
rms_error = data[:,2]

log_N = np.log(N)
log_Rn2 = np.log(Rn_squared)
slope, intercept = np.polyfit(log_N,log_Rn2, 1)
plt.errorbar(log_N, log_Rn2, yerr = rms_error/Rn_squared, fmt='o', label=r'log( $\langle R_N^2 \rangle$) vs log(N)', capsize=5)
plt.plot(log_N, slope * log_N + intercept, color='red', label=f'Fit: slope = {slope:.5f}')
plt.xlabel(r'log(N)')
plt.ylabel(r'log( $\langle R_N^2 \rangle$)')
plt.title(r'Log-Log Plot of $\langle R_N^2 \rangle$ vs N for Self Avoiding Walk where sample size S=100000')
plt.legend()
plt.tight_layout()
plt.show()
```
