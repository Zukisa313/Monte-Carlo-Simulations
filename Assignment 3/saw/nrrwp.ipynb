```python
import matplotlib.pyplot as plt

# Read the coordinates from the file
coordinates = []
with open("saw_coordinates.txt", "r") as file:
    for line in file:
        x, y = map(int, line.split())
        coordinates.append((x, y))

# Unzip the coordinates into two lists: x and y
x_vals, y_vals = zip(*coordinates)

# Plot the path
plt.figure(figsize=(8, 8))
plt.plot(x_vals, y_vals, marker='o', color='b', linestyle='-')

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Self-Avoiding Walk Path')

# Mark the starting and ending points
plt.scatter(x_vals[0], y_vals[0], color='green', label='Start', zorder=5)
plt.scatter(x_vals[-1], y_vals[-1], color='red', label='End', zorder=5)

# Display the legend and grid
plt.legend()
plt.grid(True)
plt.show()
```
