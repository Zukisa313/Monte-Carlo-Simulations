```python
import numpy as np
import matplotlib.pyplot as plt

# Load data
filename = 'SRWDATA3.txt' 
data = np.loadtxt(filename, delimiter=' ')  

# Extract columns
N = data[:, 0]
Rn_squared = data[:, 1]
Rn = data[:, 2]
rms_error = data[:, 3] 
rms_error_Rn = data[:, 4]


# Compute log values
log_N = np.log(N)
log_Rn2 = np.log(Rn_squared)
log_Rn = np.log(Rn)
# Perform linear fit
slope, intercept = np.polyfit(log_N, log_Rn2, 1)
slope2 , intercept2 = np.polyfit(log_N, log_Rn, 1)

# Plotting
plt.figure(figsize=(12, 6))
plt.errorbar(log_N, log_Rn2, yerr=rms_error / Rn_squared, fmt='o', label=r'log( $\langle R_N^2 \rangle$) vs log(N)', capsize=5)
plt.plot(log_N, slope * log_N + intercept, color='red', label=f'Fit: slope = {slope:.2f}')
plt.xlabel(r'log(N)')
plt.ylabel(r'log( $\langle R_N^2 \rangle$)')
plt.title(r'Log-Log Plot of $\langle R_N^2 \rangle$ vs N for Simple Random Walk where sample size S=20000')
plt.legend()
plt.tight_layout()
plt.show()

plt.errorbar(log_N, log_Rn, yerr=rms_error_Rn / Rn, fmt='o', label=r'log( $\langle R_N \rangle$) vs log(N)', capsize=5)
plt.plot(log_N, slope2 * log_N + intercept2, color='red', label=f'Fit: slope = {slope2:.2f}')
plt.xlabel(r'log(N)')
plt.ylabel(r'log( $\langle R_N \rangle$)')
plt.title(r'Log-Log Plot of $\langle R_N \rangle$ vs N for Simple Random Walk where sample size S=20000')
plt.legend()
plt.tight_layout()
plt.show()
```
